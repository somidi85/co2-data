# CO2 data
In this notebook, we explore the CO2 data from the repository related to Our World In Data.


```python
import pandas as pd 

data_df = pd.read_csv("owid-co2-data.csv")

data_df.head()
```

```python
data_df.info()
```

```python
from matplotlib import pyplot as plt

%matplotlib inline

fig, ax = plt.subplots(figsize=(15,7))
data_df.groupby("year")["co2"].sum().plot(ax=ax)
```

```python
selected_columns = ["gas_co2", "cement_co2", "trade_co2", "oil_co2", "coal_co2", "consumption_co2"]

fig, ax = plt.subplots(figsize=(15, 7))
df = data_df.groupby("year")[selected_columns].sum()

df.plot(ax = ax)
```
